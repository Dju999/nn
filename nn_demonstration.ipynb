# Linear regression

```python
num_train_experiment = 5
num_batches=1000
losses = np.array([])
batch_mark = np.array([])
acc = np.array([])
for exp in range(num_train_experiment):
    loss, avg_acc = nn_lib.linear_regression_mnist(nmist, num_batches=num_batches)
    losses = np.append(losses, loss)
    batch_mark = np.append(batch_mark, range(num_batches))
    acc = np.append(acc, avg_acc)
# plot data
ax = plt.axes()
sns.regplot(batch_mark, losses, x_ci='sd', lowess=True, scatter_kws={"s": 0.05})
ax.set_title('Avg accuracy={:.3f}'.format(avg_acc.mean()))
plt.show()
```
